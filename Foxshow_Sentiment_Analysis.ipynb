```python
import Scrapset as m
```

```python
import pandas as pd
import numpy as np
```

```python
get_title = pd.read_csv("merge.csv")
```

```python
link = get_title['tconst']
new_link = []
for i in list(link):
    new_link_i = "https://www.imdb.com/title/"+str(i)+"/reviews"
    new_link.append(new_link_i)
```

```python
for i in new_link:
    print(i)
```

```python
df = m.imdb()
data_test = []
for i in new_link:
    data_test_i = df.comments(i)
    data_test.append(data_test_i)
```

```python
data_test_2 = pd.DataFrame(data_test)
```

```python
data_test_2.info()
```

```python
data_test_2['title'] = get_title['Series Title']
```

```python
data_test_2
```

```python
cleandata=[]
spoilers=0
for j in range(0,12983):
    for num, row in enumerate(data_test_2.iloc[:, j]):
        reviewdict={}
        i=0
        if row != None:
            # Necessary because there are some reviews without scores
            score=row.split("\n")[i].split("/")[0]
            try:
                score=int(score)
            except:
                i=-1
                score=np.NaN

            review = row.split("\n")[i+3:]
            # This is necessary, because the review is hidden, when it cointains spoilers.
            if "Warning: Spoilers" in review:
                spoilers+=1
                continue
            
            
            title=row.split("\n")[i+1]
            username=row.split("\n")[i+2][0:-13] # doable because all of these are from 2023 July
            date=row.split("\n")[i+2][-13:] # doable because all of these are from 2023 July
            
            fullreview=[]
            for line in review:
                if ("permalink" not in line.lower()) and ("found this helpful" not in line.lower()): fullreview.append(line)
            fullreview="".join(fullreview)

            reviewdict["score"]=score
            reviewdict["title"]=title
            reviewdict["username"]=username
            reviewdict["date"]=date
            reviewdict["review"]=fullreview
            reviewdict["series_title"]=data_test_2.iloc[num, 12984]

            cleandata.append(reviewdict)
```

```python
clean_df=pd.DataFrame.from_dict(cleandata)
clean_df=clean_df.drop_duplicates()
```

```python
clean_df.info()
```

```python
clean_df = clean_df.sort_values('series_title')
clean_df.head(40)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#VADER (Valence Aware Dictionary and sEntiment Reasoner): This is a pre-trained sentiment analysis tool specifically designed for social media text. It's available through the nltk library and is particularly good at understanding sentiment in informal language.
def vader_score(x):
    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(x)
    return sentiment['compound']

def analyze(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Calculate VADER sentiment scores and analysis labels
clean_df['score'] = clean_df['review'].apply(vader_score)
clean_df['analysis'] = clean_df['score'].apply(analyze)

clean_df.head()
#here analysis is done for text colume and the data is stored in analysis column

#In this example, we've assumed that a sentiment score greater than or equal to 0.05 is considered positive, a score less than or equal to -0.05 is considered negative, and scores in between are considered neutral. You can adjust these thresholds based on your specific needs.
#Remember to adjust the analyze function according to your specific requirements for labeling sentiment scores. Also, make sure you have the NLTK library installed to use the VADER sentiment analyzer.
```
